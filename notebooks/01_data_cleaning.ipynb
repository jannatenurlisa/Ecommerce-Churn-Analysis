# 01. Data cleaning

## Load and prepare the raw dataset for analysis

## Work preformed
1. Loaded raw CSV (74,817 rows, 7 columns)
2. Converted the 'TimeStamp' from str to datetime
3. Droped 'Outcome' column (redundant/ repetive with 'EventType')
4. Filled 'ProductID' null cells with 'N/A' for str type 
5. Filled 'Amount' null cells with 0 for int type 
6. Sort by userID and timestamp
7. Saved cleaned file to date/cleaned/

## Output
- Cleaned dataset: 74,817, 6 columns
- Date range: 2024-01-01 to 2024-07-24
- Zero null values remaining

```python
import pandas as pd 
import numpy as np

```

```python
#dataset load and read
df = pd.read_csv(r'F:\Projects\User Retention Study\User Churn Analysis\Data\raw\ecommarce_clickstream_transactions.csv')
```

```python
#Knowing the dataset better
print("Dataset Shape:", df.shape)
print("\nFirst few rows:")
print(df.head())
print("\nColumn names and types")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
print("\nBasic status:")
print(df.describe())
```

```python
#checking Outcome and eventType column

print(df['Outcome'].value_counts(dropna=False))
print("\nUnique EventTypes:")
print(df['EventType'].value_counts())
```

```python
#converting the 'TimeStamp' from str to datetime

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

#check
print(df['Timestamp'].dtype)
print(df['Timestamp'].head())
```

```python
#drop 'Outcome' column (redundant/ repetive with 'EventType')

df = df.drop(columns=['Outcome'])

#check
print(df.columns)
print(df.shape)
```

```python
#Fill null cells with 'N/A' for str type

df['ProductID'] = df['ProductID'].fillna('N/A')

#Fill null cells with '0' for int type
df['Amount'] = df['Amount'].fillna(0)

#check
print(df.isnull().sum())
```

```python
#sort by userID and timestamp
df = df.sort_values(by=['UserID', 'Timestamp']).reset_index(drop=True)


#check
print(df.head(10))
```

```python
#save the cleaned dataset
df.to_csv(r'F:\Projects\User Retention Study\User Churn Analysis\Data\cleaned\ecommerce_cleaned.csv', index=False)

#check
print("File saved successfully")
print(f"Final shape: {df.shape}")
print(f"Date range: {df['Timestamp'].min()} to {df['Timestamp'].max()}")
```
